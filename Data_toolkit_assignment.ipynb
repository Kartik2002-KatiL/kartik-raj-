# a. The difference in np.array, np.asarray and np.asanyarray

In NumPy, `np.array`, `np.asarray`, and `np.asanyarray` are functions used to create arrays, but they differ in how they handle input types and data copying:

1. **`np.array`**:
   - Creates a new NumPy array from any object, including lists, tuples, or other array-like structures.
   - By default, it copies the data, but you can use the `copy` parameter to specify whether to make a copy or not.
   - You can also specify the data type using the `dtype` parameter.
   - Example:
     ```python
     import numpy as np
     arr = np.array([1, 2, 3])  # Creates a new array with a copy of the data
     ```

2. **`np.asarray`**:
   - Similar to `np.array`, but it does not copy the data if the input is already a NumPy array (i.e., it returns the original array).
   - This is useful for avoiding unnecessary data copies when you want to ensure the input is converted to an array type but don't need a new copy.
   - Example:
     ```python
     arr1 = np.array([1, 2, 3])
     arr2 = np.asarray(arr1)  # No copy, arr2 is arr1
     ```

3. **`np.asanyarray`**:
   - Similar to `np.asarray`, but it allows input of any array type (e.g., `ndarray`, `masked_array`, etc.) and will convert it to the most generic array type.
   - If the input is already an array-like object, it returns it as is without making a copy. However, if it's not an array, it converts it to an `ndarray`.
   - Example:
     ```python
     arr1 = np.array([1, 2, 3])
     arr2 = np.asanyarray(arr1)  # Returns arr1 unchanged
     arr3 = np.asanyarray([[1, 2], [3, 4]])  # Returns a 2D array
     ```

### Summary:
- Use `np.array` when you want a new array and possibly a copy.
- Use `np.asarray` to convert to an array without copying if the input is already an array.
- Use `np.asanyarray` to convert to the most generic array type while allowing for different array types without making unnecessary copies.

# Q14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:

# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity')

# b) Overlay a line plot representing the normal distribution's probability density function (PDF)

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array with 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Generate x values for the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

# Calculate the PDF using the mean and standard deviation of the data
mean = np.mean(data)
std = np.std(data)
p = norm.pdf(x, mean, std)

# Overlay the PDF as a line plot
plt.plot(x, p, 'k', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normal Distribution with PDF')

# Display the plot
plt.show()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
quadrant = np.where(x >= 0, np.where(y >= 0, 1, 4), np.where(y >= 0, 2, 3))

# Create the scatter plot with Seaborn, coloring points by quadrant
sns.scatterplot(x=x, y=y, hue=quadrant, palette=['red', 'blue', 'green', 'purple'])

# Add a legend
plt.legend(title='Quadrant')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the plot
plt.show()
````

# Q18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'8.

# Q19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'8

````python
!pip install bokeh
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool
# Import factor_cmap from bokeh.transform
from bokeh.transform import factor_cmap
import numpy as np
import pandas as pd

output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 20, size=5)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add a vbar glyph
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color=factor_cmap('categories', palette=['red', 'green', 'blue', 'orange', 'purple'], factors=categories))


# Add hover tooltips
hover = HoverTool(tooltips=[
    ("Category", "@categories"),
    ("Value", "@values")
])
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Category'
p.yaxis.axis_label = 'Value'

# Show the plot
show(p)
````
