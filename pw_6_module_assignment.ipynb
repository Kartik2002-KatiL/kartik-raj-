```python
#3. Create a dataset representing categories and their corresponding values. Compare different categories
# based on numerical values.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create a sample dataset
data = {'Category': ['A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
        'Value': [10, 12, 8, 9, 11, 5, 7, 6, 8]}
df = pd.DataFrame(data)

# Box plot to compare categories based on values
sns.boxplot(x='Category', y='Value', data=df)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot Comparison of Categories')
plt.show()

# Violin plot to compare categories based on values
sns.violinplot(x='Category', y='Value', data=df)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Violin Plot Comparison of Categories')
plt.show()

# Bar plot to compare categories based on the mean values
mean_values = df.groupby('Category')['Value'].mean()
plt.bar(mean_values.index, mean_values.values)
plt.xlabel('Categories')
plt.ylabel('Mean Value')
plt.title('Bar Plot Comparison of Categories (Mean Values)')
plt.show()
```

```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
# variable across different categories.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataset
data = {'Category': ['A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'C'],
        'Value': [10, 12, 8, 9, 11, 5, 7, 6, 8]}
df = pd.DataFrame(data)

# Visualize the distribution of 'Value' across different 'Category' using a box plot
sns.boxplot(x='Category', y='Value', data=df)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Distribution of Value across Categories')
plt.show()

# You can also use a violin plot to visualize the distribution
sns.violinplot(x='Category', y='Value', data=df)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Distribution of Value across Categories (Violin Plot)')
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
# heatmap.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic data with correlated features
np.random.seed(42)
num_samples = 100
x1 = np.random.randn(num_samples)
x2 = 0.7 * x1 + np.random.randn(num_samples) * 0.3
x3 = -0.5 * x1 + np.random.randn(num_samples) * 0.4
data = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

# Install Bokeh if not already installed
!pip install bokeh

# Import necessary libraries
from bokeh.plotting import figure, output_notebook, show
import numpy as np

# Enable notebook output
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add a line glyph to the figure
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Show the plot
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
# markers based on the 'sizes' and 'colors' columns.

# Generate random data
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.io import output_notebook

output_notebook()

N = 100
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(10, 20, size=N)
colors = ["#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(np.floor(50+2*x*200), np.floor(30+2*y*200))]

# Create a Bokeh figure
p = figure(title="Scatter Plot with Bokeh", x_axis_label='x', y_axis_label='y')

# Add a scatter glyph to the figure with different marker sizes and colors
p.scatter(x, y, size=sizes, color=colors, alpha=0.8)

# Show the plot
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add a vbar glyph for the bar chart
p.vbar(x=fruits, top=counts, width=0.9)

# Customize the plot
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

# Generate the histogram data - this line was previously commented out causing the error
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram with Bokeh", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph for the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```
