# 2.Describe the role of predefined key words and provide example of how they are used in a program.
###Ans.Predefined keywords in Python are reserved words that have special meaning and serve specific functions in the language. These keywords cannot be used as identifiers (like variable names) because they are integral to Python’s syntax and control structures.

### Key Roles of Predefined Keywords:

1. **Control Flow**: Keywords like `if`, `elif`, `else`, `for`, `while`, and `break` help control the flow of execution in a program.
  
2. **Data Definition**: Keywords such as `def`, `class`, and `import` are used to define functions, classes, and import modules, respectively.

3. **Variable Management**: Keywords like `global` and `nonlocal` are used to declare variable scopes.

4. **Exception Handling**: Keywords like `try`, `except`, `finally`, and `raise` are used for handling exceptions and errors gracefully.

### Examples of Usage:

1. **Control Flow Example**:
   ```python
   age = 18
   if age >= 18:
       print("You are an adult.")
   else:
       print("You are a minor.")
   ```

   Here, `if`, `else`, and `print` are keywords. The `if` keyword checks a condition, and the `else` keyword provides an alternative action.

2. **Function Definition Example**:
   ```python
   def greet(name):
       print(f"Hello, {name}!")

   greet("Alice")
   ```

   In this example, `def` is used to define a function named `greet`, which takes a parameter `name`.

3. **Loop Example**:
   ```python
   for i in range(5):
       print(i)
   ```

   The `for` keyword creates a loop that iterates over a range of numbers from 0 to 4.

4. **Exception Handling Example**:
   ```python
   try:
       x = 10 / 0
   except ZeroDivisionError:
       print("You can't divide by zero!")
   ```

   Here, `try` begins a block that may raise an exception, and `except` catches that specific exception.

# 3.Compare and contrast mutable and immutable objects in Python with example.
###Ans.In Python, objects can be classified as mutable or immutable based on whether their state can be changed after creation. Here’s a detailed comparison:

### Mutable Objects

**Definition**: Mutable objects can be modified after they are created. This means you can change their contents without creating a new object.

**Examples**:
- Lists
- Dictionaries
- Sets

**Characteristics**:
- Can be changed in place.
- The original object retains its identity (the same object can be referenced).
- Methods like `append()`, `extend()`, `remove()`, etc., are commonly used.

**Example**:
```python
# Mutable example: List
my_list = [1, 2, 3]
print("Original list:", my_list)

my_list.append(4)  # Modifying the list
print("Modified list:", my_list)
```
**Output**:
```
Original list: [1, 2, 3]
Modified list: [1, 2, 3, 4]
```

### Immutable Objects

**Definition**: Immutable objects cannot be modified after they are created. Any operation that tries to change the object results in the creation of a new object.

**Examples**:
- Tuples
- Strings
- Frozensets

**Characteristics**:
- Cannot be changed in place.
- A new object is created for any modification, leading to a new identity.
- Common methods like `join()`, `replace()`, etc., create and return new strings.

**Example**:
```python
# Immutable example: String
my_string = "Hello"
print("Original string:", my_string)

# Trying to modify the string (will create a new string instead)
new_string = my_string.replace("H", "J")
print("Modified string:", new_string)
```
**Output**:
```
Original string: Hello
Modified string: Jello

# 4.Discuss the different type of operators in Python and provide example of how they are used.
###Ans.Python supports various types of operators that perform specific operations on values and variables. Here’s a breakdown of the different types of operators in Python, along with examples:

### 1. Arithmetic Operators
These operators perform basic mathematical operations.

| Operator | Description          | Example          |
|----------|----------------------|------------------|
| `+`      | Addition             | `a + b`          |
| `-`      | Subtraction          | `a - b`          |
| `*`      | Multiplication       | `a * b`          |
| `/`      | Division             | `a / b`          |
| `//`     | Floor Division       | `a // b`         |
| `%`      | Modulus (remainder)  | `a % b`          |
| `**`     | Exponentiation       | `a ** b`         |

**Example**:
```python
a = 10
b = 3
print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.3333...
print(a // b) # 3
print(a % b)  # 1
print(a ** b) # 1000
```

### 2. Comparison Operators
These operators compare values and return a Boolean result (`True` or `False`).

| Operator | Description       | Example        |
|----------|-------------------|----------------|
| `==`     | Equal to          | `a == b`      |
| `!=`     | Not equal to      | `a != b`      |

| `>=`     | Greater than or equal to | `a >= b`  |
| `<=`     | Less than or equal to    | `a <= b`  |

**Example**:
```python
a = 10
b = 20
print(a == b)  # False
print(a != b)  # True
print(a > b)   # False
print(a < b)   # True
print(a >= b)  # False
print(a <= b)  # True
```

| Operator | Description      | Example         |
|----------|------------------|-----------------|

| `not`    | Logical NOT      | `not a`         |

**Example**:
```python
a = True
b = False
print(a and b)  # False
print(a or b)   # True
print(not a)    # False
```

### 4. Bitwise Operators
These operators perform operations on binary representations of integers.

| Operator | Description       | Example        |
|----------|-------------------|----------------|
| `&`      | Bitwise AND       | `a & b`        |
| `|`      | Bitwise OR        | `a | b`        |
| `^`      | Bitwise XOR       | `a ^ b`        |
| `~`      | Bitwise NOT       | `~a`           |
| `<<`     | Left Shift        | `a << 2`       |
| `>>`     | Right Shift       | `a >> 2`       |

**Example**:
```python
a = 5  # 0101 in binary
b = 3  # 0011 in binary
print(a & b)  # 1  (0001)
print(a | b)  # 7  (0111)
print(a ^ b)  # 6  (0110)
print(~a)     # -6 (inverts the bits)
print(a << 1) # 10 (1010)
print(a >> 1) # 2  (0010)
```

# 5.Explain the concept of type casting with example.
###Ans.Type casting in Python refers to the process of converting one data type into another. This is useful when you need to perform operations between different types or when you want to ensure a variable is of a specific type.

### Types of Type Casting

1. **Implicit Casting (Automatic Type Conversion)**:
   Python automatically converts one data type to another when it makes sense to do so, such as during arithmetic operations.

   **Example**:
   ```python
   a = 5          # an integer
   b = 2.5        # a float
   c = a + b      # Python converts 'a' to float
   print(c)       # Output: 7.5
   ```

2. **Explicit Casting (Manual Type Conversion)**:
   This involves using built-in functions to convert data types. Python provides several functions for this purpose, such as `int()`, `float()`, `str()`, and `list()`.

   **Examples**:

   - **Converting to Integer**:
     ```python
     num_str = "10"
     num_int = int(num_str)  # Convert string to integer
     print(num_int)           # Output: 10
     ```

   - **Converting to Float**:
     ```python
     num_str = "3.14"
     num_float = float(num_str)  # Convert string to float
     print(num_float)             # Output: 3.14
     ```

   - **Converting to String**:
     ```python
     num = 100
     num_str = str(num)  # Convert integer to string
     print(num_str)      # Output: "100"
     ```

   - **Converting to List**:
     ```python
     my_string = "hello"
     my_list = list(my_string)  # Convert string to list of characters
     print(my_list)              # Output: ['h', 'e', 'l', 'l', 'o']
     ```

### Summary
Type casting allows for greater flexibility in programming by enabling you to convert between different data types as needed. Implicit casting happens automatically in compatible situations, while explicit casting requires using specific functions to convert types manually. Understanding type casting is essential for effective data manipulation and operations in Python.

# 6.How do conditional statement work in Python . Illustrate with example.
###Ans.Conditional statements in Python allow you to execute certain pieces of code based on whether a condition evaluates to `True` or `False`. The primary conditional statements are `if`, `elif`, and `else`. Here’s a breakdown of how they work, along with examples.

### Basic Structure

1. **if Statement**: Evaluates a condition. If it’s `True`, the block of code within the `if` statement runs.
2. **elif Statement**: Stands for "else if." You can check multiple conditions sequentially.
3. **else Statement**: Runs if none of the preceding conditions are `True`.

### Syntax

```python
if condition1:
    # Code to execute if condition1 is True
elif condition2:
    # Code to execute if condition2 is True
else:
    # Code to execute if neither condition is True
```

### Example

Let’s say we want to determine if a person is eligible to vote based on their age.

```python
age = 20

if age >= 18:
    print("You are eligible to vote.")
elif age < 0:
    print("Age cannot be negative.")
else:
    print("You are not eligible to vote yet.")
```

### Explanation

1. **if age >= 18**:
   - Checks if `age` is greater than or equal to 18. If this condition is `True`, it prints "You are eligible to vote."
  
2. **elif age < 0**:
   - This checks if `age` is negative. If this condition is `True`, it prints "Age cannot be negative." This acts as an additional check to ensure valid age input.

3. **else**:
   - If neither of the above conditions is met, it executes this block, printing "You are not eligible to vote yet."

### Another Example with Multiple Conditions

You can also use logical operators to combine multiple conditions. Here’s an example of grading based on scores:

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
```

### Explanation

1. **if score >= 90**:
   - If the score is 90 or above, it prints "Grade: A."

2. **elif score >= 80**:
   - If the score is between 80 and 89, it prints "Grade: B."

3. **Additional `elif` statements** check for grades down to "F" for scores below 60.

### Summary

Conditional statements are fundamental in programming, allowing for decision-making based on dynamic conditions. They enable you to control the flow of execution based on the values of variables or results of expressions, leading to more complex and interactive programs.

# 7.Describe the different types of loops in Python and their use cases with example.
###Ans.In Python, loops are used to execute a block of code repeatedly, either a specific number of times or until a certain condition is met. The two primary types of loops in Python are **`for` loops** and **`while` loops**. Here's a detailed explanation of each, along with examples and use cases.

### 1. `for` Loop

**Description**: The `for` loop is used to iterate over a sequence (like a list, tuple, dictionary, set, or string) or any iterable object. It executes the block of code for each item in the sequence.

**Use Cases**:
- Iterating through lists or tuples.
- Processing items in a collection.
- Performing operations on elements of a string.

**Syntax**:
```python
for variable in iterable:
    # Code to execute
```

**Example**:
```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I like {fruit}.")
```

**Output**:
```
I like apple.
I like banana.
I like cherry.
```

### 2. `while` Loop

**Description**: The `while` loop continues to execute as long as a specified condition is `True`. It's useful when the number of iterations is not known beforehand and depends on dynamic conditions.

**Use Cases**:
- Repeating actions until a certain condition changes.
- Waiting for user input or external conditions.
- Implementing algorithms that require conditional repetition.

**Syntax**:
```python
while condition:
    # Code to execute
```

**Example**:
```python
count = 0

while count < 5:
    print("Count is:", count)
    count += 1  # Increment the count
```

**Output**:
```
Count is: 0
Count is: 1
Count is: 2
Count is: 3
Count is: 4
```

### 3. Nested Loops

**Description**: You can use loops inside other loops. This is useful for iterating over multi-dimensional data structures (like lists of lists).

**Use Cases**:
- Processing matrices or multi-dimensional arrays.
- Generating combinations or permutations.

**Example**:
```python
for i in range(3):  # Outer loop
    for j in range(2):  # Inner loop
        print(f"i = {i}, j = {j}")
```

**Output**:
```
i = 0, j = 0
i = 0, j = 1
i = 1, j = 0
i = 1, j = 1
i = 2, j = 0
i = 2, j = 1
```

### 4. Loop Control Statements

Python also provides control statements to alter the flow of loops:

- **`break`**: Terminates the loop and transfers control to the statement immediately following the loop.
- **`continue`**: Skips the current iteration and moves to the next iteration of the loop.
- **`pass`**: Does nothing; it's a placeholder that can be useful when a statement is required syntactically.

**Example using `break` and `continue`**:
```python
for num in range(10):
    if num == 5:
        break  # Exit the loop when num is 5
    print(num)
```

**Output**:
```
0
1
2
3
4
```

**Example using `continue`**:
```python
for num in range(10):
    if num % 2 == 0:
        continue  # Skip even numbers
    print(num)
```

**Output**:
```
1
3
5
7
9
```

### Summary

- **`for` Loop**: Best for iterating over known sequences or collections.
- **`while` Loop**: Ideal for situations where the number of iterations is not predetermined.
- **Nested Loops**: Useful for multi-dimensional data structures.
- **Control Statements**: Provide additional flexibility in managing loop execution.

Understanding these looping constructs is crucial for writing efficient and effective Python programs.
